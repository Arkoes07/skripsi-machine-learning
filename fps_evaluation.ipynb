```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Performa FPS berdasarkan metode pengambilan fitur

```python
df = pd.read_csv('fps_test.csv')
```

```python
fig, axis = plt.subplots(2,1, figsize=(8,6))

sns.lineplot(x='time', y='frame', hue='Metode', data=df, ax=axis[0])
sns.lineplot(x='time', y='FPS', hue='Metode', data=df, ax=axis[1])
```

```python
df['Metode'][:7]
```

```python
order = ['tanpa pelacakan','pelacakan 5 frame','pelacakan 10 frame','pelacakan 15 frame','pelacakan 20 frame','pelacakan 25 frame']
bplot = sns.barplot(x='Metode', y='FPS', data=df.groupby('Metode').mean().reset_index(), order=order)
plt.ylim(0,15)
plt.rcParams["figure.figsize"] = (11,3)
plt.title("FPS Rata-rata Setiap Metode")
for p in bplot.patches:
     bplot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.savefig('plots/FPS_comparison.png')
```
