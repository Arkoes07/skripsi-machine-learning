```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Performa FPS berdasarkan metode pengambilan fitur

```python
df = pd.read_csv('fps_test.csv')
```

```python
df.groupby('method').mean()
```

```python
fig, axis = plt.subplots(3,1, figsize=(12,12))

sns.lineplot(x='time', y='frame', hue='method', data=df, ax=axis[0])
sns.lineplot(x='time', y='fps', hue='method', data=df, ax=axis[1])
bplot = sns.barplot(x='method', y='fps', data=df.groupby('method').mean().reset_index(), ax=axis[2])
# for p in bplot.patches:
#     bplot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```
